```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure seaborn style is set
sns.set_style('whitegrid')

# Read the CSV file
df = pd.read_csv('/content/Base model results.csv')

# Create a figure with subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 16))

# 1. Training Losses Plot
ax1.plot(df['epoch'], df['train/box_loss'], label='Box Loss', linewidth=2)
ax1.plot(df['epoch'], df['train/cls_loss'], label='Classification Loss', linewidth=2)
ax1.plot(df['epoch'], df['train/dfl_loss'], label='DFL Loss', linewidth=2)
ax1.set_title('Training Losses Over Epochs', fontsize=14)
ax1.set_xlabel('Epoch', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.legend()
ax1.grid(True)

# 2. Validation Metrics Plot
ax2.plot(df['epoch'], df['metrics/precision(B)'], label='Precision', linewidth=2)
ax2.plot(df['epoch'], df['metrics/recall(B)'], label='Recall', linewidth=2)
ax2.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP50', linewidth=2)
ax2.plot(df['epoch'], df['metrics/mAP50-95(B)'], label='mAP50-95', linewidth=2)
ax2.set_title('Validation Metrics Over Epochs', fontsize=14)
ax2.set_xlabel('Epoch', fontsize=12)
ax2.set_ylabel('Score', fontsize=12)
ax2.legend()
ax2.grid(True)

# 3. Learning Rate Schedule
ax3.plot(df['epoch'], df['lr/pg0'], label='Learning Rate', linewidth=2, color='purple')
ax3.set_title('Learning Rate Schedule', fontsize=14)
ax3.set_xlabel('Epoch', fontsize=12)
ax3.set_ylabel('Learning Rate', fontsize=12)
ax3.legend()
ax3.grid(True)

# 4. Validation Losses Plot
ax4.plot(df['epoch'], df['val/box_loss'], label='Val Box Loss', linewidth=2)
ax4.plot(df['epoch'], df['val/cls_loss'], label='Val Class Loss', linewidth=2)
ax4.plot(df['epoch'], df['val/dfl_loss'], label='Val DFL Loss', linewidth=2)
ax4.set_title('Validation Losses Over Epochs', fontsize=14)
ax4.set_xlabel('Epoch', fontsize=12)
ax4.set_ylabel('Loss', fontsize=12)
ax4.legend()
ax4.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()

# Additional detailed plots

# 5. Precision vs Recall Plot
plt.figure(figsize=(10, 6))
plt.plot(df['metrics/recall(B)'], df['metrics/precision(B)'], '-o', markersize=2)
plt.title('Precision vs Recall Curve', fontsize=14)
plt.xlabel('Recall', fontsize=12)
plt.ylabel('Precision', fontsize=12)
plt.grid(True)
plt.show()

# 6. mAP Progress Plot
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP50', linewidth=2)
plt.plot(df['epoch'], df['metrics/mAP50-95(B)'], label='mAP50-95', linewidth=2)
plt.title('mAP Progress Over Epochs', fontsize=14)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('mAP', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV files
base_model_df = pd.read_csv('/content/Base model results.csv')
custom_model_df = pd.read_csv('/content/results.csv')
# Set Seaborn style for consistency
sns.set_style('whitegrid')

# Create a figure with subplots for each metric
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle("Comparison of Base YOLO vs Customized YOLO (Ghost-CBAM)", fontsize=16)

# 1. Training Losses
axes[0, 0].plot(base_model_df['epoch'], base_model_df['train/box_loss'], label='Base Box Loss', linewidth=2)
axes[0, 0].plot(custom_model_df['epoch'], custom_model_df['train/box_loss'], label='Custom Box Loss', linewidth=2, linestyle='--')
axes[0, 0].plot(base_model_df['epoch'], base_model_df['train/cls_loss'], label='Base Class Loss', linewidth=2)
axes[0, 0].plot(custom_model_df['epoch'], custom_model_df['train/cls_loss'], label='Custom Class Loss', linewidth=2, linestyle='--')
axes[0, 0].set_title("Training Loss Comparison")
axes[0, 0].set_xlabel("Epoch")
axes[0, 0].set_ylabel("Loss")
axes[0, 0].legend()
axes[0, 0].grid(True)

# 2. Validation Losses
axes[0, 1].plot(base_model_df['epoch'], base_model_df['val/box_loss'], label='Base Val Box Loss', linewidth=2)
axes[0, 1].plot(custom_model_df['epoch'], custom_model_df['val/box_loss'], label='Custom Val Box Loss', linewidth=2, linestyle='--')
axes[0, 1].plot(base_model_df['epoch'], base_model_df['val/cls_loss'], label='Base Val Class Loss', linewidth=2)
axes[0, 1].plot(custom_model_df['epoch'], custom_model_df['val/cls_loss'], label='Custom Val Class Loss', linewidth=2, linestyle='--')
axes[0, 1].set_title("Validation Loss Comparison")
axes[0, 1].set_xlabel("Epoch")
axes[0, 1].set_ylabel("Loss")
axes[0, 1].legend()
axes[0, 1].grid(True)

# 3. Precision vs Recall
axes[0, 2].plot(base_model_df['metrics/recall(B)'], base_model_df['metrics/precision(B)'], '-o', markersize=2, label='Base Precision vs Recall')
axes[0, 2].plot(custom_model_df['metrics/recall(B)'], custom_model_df['metrics/precision(B)'], '-o', markersize=2, label='Custom Precision vs Recall', linestyle='--')
axes[0, 2].set_title("Precision vs Recall Curve")
axes[0, 2].set_xlabel("Recall")
axes[0, 2].set_ylabel("Precision")
axes[0, 2].legend()
axes[0, 2].grid(True)

# 4. mAP Progress (mAP50 and mAP50-95)
axes[1, 0].plot(base_model_df['epoch'], base_model_df['metrics/mAP50(B)'], label='Base mAP50', linewidth=2)
axes[1, 0].plot(custom_model_df['epoch'], custom_model_df['metrics/mAP50(B)'], label='Custom mAP50', linewidth=2, linestyle='--')
axes[1, 0].plot(base_model_df['epoch'], base_model_df['metrics/mAP50-95(B)'], label='Base mAP50-95', linewidth=2)
axes[1, 0].plot(custom_model_df['epoch'], custom_model_df['metrics/mAP50-95(B)'], label='Custom mAP50-95', linewidth=2, linestyle='--')
axes[1, 0].set_title("mAP Progress Comparison")
axes[1, 0].set_xlabel("Epoch")
axes[1, 0].set_ylabel("mAP")
axes[1, 0].legend()
axes[1, 0].grid(True)

# 5. Learning Rate Schedule Comparison
axes[1, 1].plot(base_model_df['epoch'], base_model_df['lr/pg0'], label='Base Learning Rate (pg0)', color='purple', linewidth=2)
axes[1, 1].plot(custom_model_df['epoch'], custom_model_df['lr/pg0'], label='Custom Learning Rate (pg0)', color='purple', linewidth=2, linestyle='--')
axes[1, 1].plot(base_model_df['epoch'], base_model_df['lr/pg1'], label='Base Learning Rate (pg1)', color='blue', linewidth=2)
axes[1, 1].plot(custom_model_df['epoch'], custom_model_df['lr/pg1'], label='Custom Learning Rate (pg1)', color='blue', linewidth=2, linestyle='--')
axes[1, 1].set_title("Learning Rate Schedule Comparison")
axes[1, 1].set_xlabel("Epoch")
axes[1, 1].set_ylabel("Learning Rate")
axes[1, 1].legend()
axes[1, 1].grid(True)

# Hide the empty subplot
axes[1, 2].axis('off')

# Adjust layout and display the plots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
